```python
df=pd.read_pickle('Purdue_ECE_Profs.pkl')

keywords_to_remove = ["my", "research", "focus", "focuses", "on"]
df['Research'] = df['Research'].apply(lambda x: ' '.join(word for word in x.split() if word.lower() not in keywords_to_remove))

df['Areas of Interest'] = df['Areas of Interest'].apply(lambda x: ", ".join(x))
df
```

```python
# df1=df[:2].copy()
# df1.iloc[0]['Email']="aryamatomar@gmail.com"
# df1.iloc[1]['Email']="cynicdoom@gmail.com"
# df1

df=df[24:]
df
```

```python
with open("bio.txt", "r") as file:
    email_template = file.read()

email = "aryamann.t@alumni.iitgn.ac.in"
subject = "Request for Master's Thesis Supervision - IIT Gandhinagar Graduate with Published Research"
filenames = ["Resume.pdf", "Transcript.pdf", "IITGN_LOR.pdf", "GSTN_LOR_SVP.pdf"]
for i, row in tqdm.tqdm(df.iterrows(),total=df.shape[0]):
    receiver_email = row['Email']
    professor_name = row['Name']
    research = row['Research']
    aoi = row['Areas of Interest']
    message = email_template.format(professor_name=professor_name, research=research, aoi=aoi)

    msg = MIMEMultipart()
    msg['From'] = email
    msg['To'] = receiver_email
    msg['Subject'] = subject
    msg.attach(MIMEText(message, 'plain'))

    for filename in filenames:
        try:
            with open(filename, "rb") as attachment:
                part = MIMEBase("application", "octet-stream")
                part.set_payload(attachment.read())
                encoders.encode_base64(part)
                part.add_header(
                    "Content-Disposition",
                    f"attachment; filename={filename}",
                )
                msg.attach(part)
        except Exception as e:
            print(f"Could not attach file {filename}. Error: {e}")

    try:
        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()  
        server.login(email, "qjcz vtbc zumb wpbd")  
        server.send_message(msg)  
        # print(f"Email has been sent to {receiver_email}")
    except Exception as e:
        print(f"Failed to send email. Error: {e}")
    finally:
        server.quit()
```
